```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sys
import os
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

# Load labels from .csv label file
labels = pd.read_csv("trainLabels.csv")
```

```python
def getNum(c):
  return int(c.split('.')[0])

sorted_files = sorted(os.listdir("train"), key=getNum)
sorted_test_files = sorted(os.listdir("test"), key=getNum)
```

```python
import matplotlib.pyplot as plt

# Define training and test sets path
train_path = "train/"
test_path = "test/"

# Identify the minimum width and height of images to resize based on these images
img = plt.imread(train_path + sorted_files[0])
min_height = img.shape[0]
min_width = img.shape[1]
height_indx = sorted_files[0]
width_indx = sorted_files[0]

for file in sorted_files:
  img = plt.imread(train_path + file)
  if img.shape[0] < min_height:
    min_height = img.shape[0]
    height_indx = file
  if img.shape[1] < min_width:
    min_width = img.shape[1]
    width_indx = file

min_height, height_indx, min_width, width_indx
```

```python
# Identified outliers in training data in terms of shape, we will try to identify and delete msot useless pictures
from scipy import stats
import numpy as np

height_shapes = []
width_shapes = []

for file in sorted_files:
  img = plt.imread(train_path + file)
  height_shapes.append(img.shape[0])
  width_shapes.append(img.shape[1])

height_Q1 = np.percentile(height_shapes, 25, interpolation = 'midpoint')
height_Q3 = np.percentile(height_shapes, 75, interpolation = 'midpoint')
height_IQR = height_Q3 - height_Q1

width_Q1 = np.percentile(width_shapes, 25, interpolation = 'midpoint')
width_Q3 = np.percentile(width_shapes, 75, interpolation = 'midpoint')
width_IQR = width_Q3 - width_Q1

n, bins, patches = plt.hist(width_shapes, bins=80)
n2, bins2, patches2 = plt.hist(height_shapes, bins=80)
plt.legend(["width", "height"])
print("width frequency and bins:\n", n,"\n", bins,"\nheight frequency and bins:\n", n2,"\n", bins2)
```

```python
import cv2
# Image preprocessing:
# we will start by removing all pictures below 7 pixels as there seems to be alot of pictures below this value
# Then we will resize each image to fit a certain size so we can input them to the CNN
# After that we normalize the data to fit a range between 0 and 1 as to help the CNN converge faster
image_list_train = []
image_list_test = []
for file in sorted_files:
  img = plt.imread(train_path + file)
  if img.shape[0] < 7 or img.shape[1] < 7:
    sorted_files.remove(file)
    labels.drop(labels.loc[labels['ID']==getNum(file)].index, axis=0, inplace=True)

# Images that only have 2 dimensions are grayscale images, therefore in order for these picture to conform
# to the number of input units for the CNN we convert these grayscale values into rgb values making it a 3 dimensional tensor  
for file in sorted_files:
  img = plt.imread(train_path + file)
  if img.ndim != 3:
    img = cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
  image_list_train.append(cv2.resize(img, (80,80)) / 255)

for file in sorted_test_files:
  img = plt.imread(test_path + file)
  if img.ndim != 3:
    img = cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
  image_list_test.append(cv2.resize(img, (80,80)) / 255)
```

```python
image_list_train = np.array(image_list_train)
image_list_train.shape

# Make a column transformer to one hot encode the Class column
ct = make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore", sparse=False), ["Class"])
)
ct.fit(labels)
labels_one_hot = ct.transform(labels)

# Make the label list so we can convert CNN output into comprehensible output
label_list = sorted(list(set(labels.Class)))
```

```python
image_list_test = np.array(image_list_test)
image_list_test.shape
```

```python
# Split the training data into 85 to 15 ratio in terms of training and validation data
split_fraction = 0.1
split_indx = int(len(image_list_train) * (1 - split_fraction))
train_set = image_list_train[:split_indx]
y_train = labels_one_hot[:split_indx]
valid_set = image_list_train[split_indx:]
y_valid = labels_one_hot[split_indx:]

len(train_set), len(valid_set), len(y_train), len(y_valid)
```

```python
# Set the seed
tf.random.set_seed(42)
lambd = 0.001
# Creating, Compiling, and fitting the model
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape=(80, 80, 3), kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2D(32,3,activation="relu", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.MaxPool2D(pool_size=2, padding='valid'),
  tf.keras.layers.Dropout(0.1),

  tf.keras.layers.Conv2D(64,3,activation="relu", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2D(64,3,activation="relu", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.MaxPool2D(pool_size=2, padding='valid'),
  tf.keras.layers.Dropout(0.2),

  tf.keras.layers.Conv2D(64,3,activation="relu", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2D(64,3,activation="relu", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.MaxPool2D(pool_size=2, padding='valid'),
  tf.keras.layers.Dropout(0.2),

  tf.keras.layers.Flatten(), 
  tf.keras.layers.Dense(256, activation="softmax", kernel_regularizer=tf.keras.regularizers.l2(lambd)),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(len(label_list), activation="softmax")
])

model.compile(loss="categorical_crossentropy", 
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0015), 
              metrics=["accuracy"])

history = model.fit(train_set,
                    y_train,
                    epochs=500,
                    validation_data=(valid_set, y_valid))
```

```python
import random
def getValue(oneHot):
  alphanumeric_indx = int(np.where(np.squeeze(oneHot) == np.amax(oneHot))[0])
  return label_list[alphanumeric_indx]

def getRandomValidationImg():
  fig, (ax1, ax2) = plt.subplots(1, 2)
  fig.suptitle('Validation img vs. Training img')

  valid_selected_img_indx = random.randint(0,len(valid_set) - 1)
  training_selected_img_indx = random.randint(0,len(train_set) - 1)

  ax1.imshow(valid_set[valid_selected_img_indx])
  ax1.title.set_text("true value: "+ 
                     getValue(y_valid[valid_selected_img_indx]) +
                     "\npredicted value: " + 
                      getValue(model.predict(tf.expand_dims(valid_set[valid_selected_img_indx], axis=0))
                              )
                    )
  ax2.imshow(train_set[training_selected_img_indx])
  ax2.title.set_text("true value: "+ getValue(y_train[training_selected_img_indx]) +
                     "\npredicted value: " + 
                      getValue(model.predict(tf.expand_dims(train_set[training_selected_img_indx], axis=0)))
                      )
```
